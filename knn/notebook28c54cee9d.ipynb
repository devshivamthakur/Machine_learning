```python
# I will use sklearn on this data so i can work on dagonsis column with it's M and B , but changed it to zero and one for more training 
# But feel free to do it with origingal data 
df = pd.get_dummies(df,{'diagnosis'})
df.head()
```

```python
# now we are going to drop some un needed columns : 
X = df.drop(['id','Unnamed: 32','diagnosis_B','diagnosis_M'
] , axis = 1)
y = df['diagnosis_B']
```

```python
# Now we are going to make train test split to split the data to train data and test data : 
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.1, random_state=42)
```

```python
# Now we are going to scale the data :
from sklearn.preprocessing import StandardScaler
```

```python
scaler = StandardScaler()
```

```python
scaler.fit(X_train)
```

```python
scaled_x_train = scaler.transform(X_train)
scaled_x_test = scaler.transform(X_test)
```

```python
# Now after scaling we will use logistic regression model : 
from sklearn.linear_model import LogisticRegression
```

```python
log_model = LogisticRegression()
```

```python
# Now we will fit the model ("train it on the X train and y train")
log_model.fit(scaled_x_train,y_train)
```

```python
# Now we are going to get y_predection : 
y_pred = log_model.predict(scaled_x_test)
y_pred
```

```python
# Now we got the y_predict and now we are going to check metrics and see our model working good or not : 
from sklearn.metrics import accuracy_score , classification_report , confusion_matrix , plot_confusion_matrix ,plot_precision_recall_curve ,plot_roc_curve
```

```python
accuracy_score(y_test,y_pred)
```

```python
print(classification_report(y_test,y_pred))
```

```python
plot_confusion_matrix(log_model,scaled_x_test,y_test)
```

```python
plot_precision_recall_curve(log_model,scaled_x_test,y_test)
```

```python
plot_roc_curve(log_model,scaled_x_test,y_test)
```

# Will use KNN Algorithm to see if the model will make better accuracy than LogisticRegrression : 


```python
# Will use KNN Algorithm to see if the model will make better accuracy : 
from sklearn.neighbors import KNeighborsClassifier
```

```python
# Will get random number for k value first 
knn_model = KNeighborsClassifier(n_neighbors=5)
```

```python
knn_model.fit(scaled_x_train,y_train)
```

```python
K_N_Y_pred = knn_model.predict(scaled_x_test)
K_N_Y_pred
```

```python
k_value5_accuracy_score = accuracy_score(y_test,K_N_Y_pred)
k_value5_accuracy_score
```

```python
# As we see the accuracy is 92% so we are going to make for loop to able to get the best K Value : 

test_error_rates = []
for k in range(1,30):
    knn_model = KNeighborsClassifier(n_neighbors=k)
    knn_model.fit(scaled_x_train , y_train)
    
    y_pred_test = knn_model.predict(scaled_x_test)
    test_error = 1- accuracy_score(y_test ,y_pred_test)
    test_error_rates.append(test_error)
```

```python
# as we see the k value when it is higher and it's error rate is lower will be very good because it will be far away from outlaier :
# So will choose 27
plt.plot(range(1,30),test_error_rates)
plt.ylabel('Error Rate')
plt.xlabel('K Neigbors')
```

```python
# so will do the model again with k_value 27 , best value for k is higher and it's error rate is low 
knn_2ndmodel = KNeighborsClassifier(n_neighbors=27)
```

```python
knn_2ndmodel.fit(scaled_x_train,y_train)
```

```python
y_pred = knn_2ndmodel.predict(scaled_x_test)
```

```python
k_value27_accuracy_score = accuracy_score(y_test,y_pred)
```
